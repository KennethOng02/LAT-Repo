# 10440-01-02-2 臺中市特殊教育學校教育經費概況(預、決算數)
https://data.gov.tw/dataset/90167

```python
df2017 = pd.read_csv('2017_01_10440-01-02-2_臺中市特殊教育學校教育經費概況(預、決算數).csv')
df2018 = pd.read_csv('2018_01_10440-01-02-2_臺中市特殊教育學校教育經費概況(預、決算數).csv')
df2017.head()
```

```python
df2017 = df2017[['ComplexName', 'Complex2Name', 'FValue']].groupby(['ComplexName', 'Complex2Name'])['FValue'].sum().reset_index()
df2017.columns = ['類型', '項目', '數額']
df2017['年份'] = 2017
df2017
```

```python
df2018 = df2018[['ComplexName', 'Complex2Name', 'FValue']].groupby(['ComplexName', 'Complex2Name'])['FValue'].sum().reset_index()
df2018.columns = ['類型', '款項', '數額']
df2018['年份'] = 2018
df2018
```

```python
frames = [df2017, df2018]

result = pd.concat(frames)
result
```
