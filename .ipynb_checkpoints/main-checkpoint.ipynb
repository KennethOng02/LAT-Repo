# 大專校院各系所僑生、港澳生數
https://data.gov.tw/dataset/26210

```python
import pandas as pd
pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv('學3-3.僑生、港澳生數及其在學比率-以「系所」統計.csv')
df.head()
```

```python
temp = df[['學年度','學校名稱', '僑生小計', '港澳生小計']]
temp
```

```python
Type = temp.groupby(['學年度', '學校名稱']).sum().reset_index()
```
